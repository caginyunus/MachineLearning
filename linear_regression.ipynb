# Notes on Linear Regression following Elements of Statistical Learning

We assume there are $p$ features and $N$ samples. The data is denoted by $N \times (p+1)$ dimensional matrix $\bX$ where the first column is a vector of all $1$s and corresponds to the constant term in a fit. $\bX$ can represented as
\begin{equation}
\bX = \begin{pmatrix} \bm x_0 & \bm x_1 & \cdots & \bm x_p \end{pmatrix},
\end{equation}
where the vectors $\bx_i$ are $N$ dimensional vectors
\begin{equation}
\bx_i = \begin{pmatrix} x_{1i} \\ \cdots \\ x_{Ni} \end{pmatrix}.
\end{equation}
The vector $\bm x_0$ is the vector of all $1$s: 
\begin{equation}
\bm x_0 = \begin{pmatrix} 1 \\ \cdots \\ 1 \end{pmatrix}.
\end{equation}
We denote the outcomes by $\by$ which is also $N$ dimensional and is given by 
\begin{equation}
\by = \begin{pmatrix} y_1 \\ \cdots \\ y_N \end{pmatrix}.
\end{equation} 
Under the assumption that the distribution of $y$ is given by (note that we are assuming the noise at $X$ and $X'$ are independent)
\begin{equation} Y = f(X_1,\cdots,X_p) + \epsilon(X),
\end{equation} where $\epsilon(X) \sim \mathcal{N}(0,\sigma_X^2)$, the log-likelihood of $\left( \bX, \by \right)$ is given by
\begin{equation} 
L \left( \bX,\by \rp = -\frac 1 2 \sum_{i=1}^N \log \lp 2\pi \sigma_X^2\rp - \frac 1 2 \sum_{i=1}^N \frac{(y_i-f(\vec x_i))^2}{\sigma_X^2},
\end{equation} 
where $\vec x_i \equiv (1, x_{i1}, \cds x_{iN})^T$ denotes the $"i"$th input. Under the assumption that $\sigma_X = \sigma$ is independent of $X$, log-likelihood is given by 
\begin{equation} 
L \left( \bX,\by \rp = -\frac N 2 \sum_{i=1}^N \log \lp 2\pi \sigma^2\rp - \frac{1}{2\sigma^2} \sum_{i=1}^N (y_i-f(\vec x_i))^2, 
\end{equation} 
Of course, one doesn't know $f(\vec x_i)$ exactly and one has candidates $f(\vec x_i; \vec  \theta)$ for it. Therefore under above assumptions, to maximize the log-likelihood one minimizes the residual sum of squares (RSS):
\begin{equation} 
\vec \theta_{LS} = \argmin_{\vec \theta} \frac{1}{N}\sum_{i=1}^{N} \left( y_i - f(\vec x_i;\vec \theta) \right)^2,
\end{equation} 
where the label 'LS' is for 'least squares'. Note that doesn't need to know $\sigma$, only that it is independent of $X$.

Next, we will assume that $f(X) = \beta_0 + \sum_{i=1}^N \b_i X_i$. Under this assumption, least square solution is given by:
$$
\begin{align} 
\hat {\bm \b} &= \argmin_{\bm \b} \left( \by - \bX \bm \b\right)^T \left( \by - \bX  \bm \b \right) \newline
&= \left( \bX^T \bX \right)^{-1} \bX^T \by
\end{align}
$$
The prediction for $\bX$ then becomes:
$$
\begin{equation} 
\hat \by = \bX \left( \bX^T \bX \right)^{-1} \bX^T \by.
\end{equation}
$$
The matrix $\bm H = \bX \inv{\left( \bX^T \bX \right)}  \bX^T$ is called the hat matrix and projects $\by$ to the subspace spanned by the columns of $\bX$. This is obvious from least squares. However, it is also instructive to directly derive the hat matrix. Consider the $QR$ decomposition of $\bX$, $\bX = \bQ\bR$ where $\bQ$ is an $N \times (p+1)$ matrix that satisfies $\bQ^T\bQ = 1$, $\bR$ is a $(p+1)\times (p+1)$ upper-triangular matrix and $span(\bQ) = span(\bX)$. Therefore the projection matrix $\bH$ is given by $\bH = \bQ\bQ^T$. Now it is easy to see that 
$$
\begin{align} 
\bQ\bQ^T &= \bX\inv{\bR}\left( \inv{\bR}\right)^T\bX^T \newline
&= \bX \inv{\left( \bR^T \bR \right)} \bX^T  \newline
&= \bX \inv{\left( \bR^T \bQ^T \bQ \bR\right)}\bX^T  \newline
&= \bX \inv{\left( \bX^T \bX \right) }\bX^T.
\end{align}
$$

It is also important to know the variance of $\hat {\bm \b}$. Assuming $\bX$ is fixed, this is easy to calculate:
$$
\begin{align}
Var(\hat {\bm\b}) &= \mathbb{E}\lk \inv{\left( \bX^T \bX \right)} \bX^T \bm \e \bm \e^T \left( \inv{\left( \bX^T \bX \right)} \bX^T\right) ^T\rk \newline 
&=\inv{\left( \bX^T \bX \right)} \bX^T \mathbb{E}\lk \bm \e \bm \e^T\rk \left( \inv{\left( \bX^T \bX \right)} \bX^T\right)^T \newline 
&= \s^2 \inv{\left( \bX^T \bX \right)} \bX^T \left( \inv{\left( \bX^T \bX \right)} \bX^T\right)^T \newline 
&= \s^2 \inv{\left( \bX^T \bX\right)},
\end{align}
$$
where we used the fact that $\mathbb{E}\lk  \bm  \e \bm \e^T \rk = \s^2 \bm{1}$. After performing the fitting, one might want to test the hypothesis $\b_i = \b^h_i$. Under this assumption, the distribution of $\hat \b_i$ is $\caln \left( \b^h_i,\s^2 \inv{\left( \bX^T \bX\right)}_{ii}\right)$. However, note that $\s^2$ is not known and we need to estimate it. An unbiased estimator is given by:
\begin{equation}
\hat{\s}^2 = \ff{1}{N-p-1}\sum_{j=1}^{N}\left( y_j - \hat y_j \right)^2,
\end{equation}
where $\hat y_j$ are predictions after fitting and is given by $\hat{\bm y} = \bm H \bm y$. The intuition behind the appearance of $N-p-1$ is as follows. While $\by$ fluctuates in an $N$-dimensional subspace, we project it into a $p+1$ dimensional subspace, therefore the fluctuations in this $p+1$ dimensional subspaces is 'overfitted' while the fluctuations in the orthogonal $N-p-1$ dimensional subspace remain. This intuition also suggests that $(N-p-1)\hat \s^2 \sim \s^2 \chi^2_{N-p-1}$. A more formal derivation is given as follows. First we express $(N-p-1)\hat \s^2$ in a more useful form:
\begin{align} 
(N-p-1)\hat \s^2 &= \left( \by - \hat \by \right)^T \left( \by - \hat \by \right) \newline  &=
\by^T \left( \bm 1 - \bm H \right)^T \left( \bm 1 - \bm H \right) \bm y \newline 
&= \by^T \left( \bm 1 - \bm H \right) \by \newline 
&= \left( {\bm f + \bm \e}\right)^T \left( \bm 1 - \bm H \right) \left( \bm f + \bm \e \right),
\end{align}
where $\bm f = \mathbb{E}\lk \bm y \rk$. By the assumption that the relation between $Y$ and $X$ is linear, it follows that $\bm f$ is in the span of $\bX$ and therefore $\bm H \bm f = \bm f$. Therefore, it follows that 
\begin{align} 
(N-p-1)\hat \s^2 = \bm \e^T \left( \bm 1 - \bm H \right) \bm \e.
\end{align} 
Since $\bH = \bH^T$, it can be diagonalized by an orthonormal matrix $\bm O$. Then (after a permutation if necessary), $\bm{\ti H} = \bm{O} \bH \bm{O}^T$ is diagonal with entries $\big( \ub{1,\cds,1 }_{p+1}, \ub{0,\cds,0}_{N-p-1}\big)$. As entries of $\bm{\e}$ are independent gaussians with variances $\s^2$, it follows that same is also true for $\bm{\ti \e} = \bm{O}\bm \e$. Then:
\begin{align} 
\ff{(N-p-1)}{\s^2}\hat \s^2 &= \ff{1}{\s^2}\sum_{a=p+2}^N \left( \ti \e_a\right)^2 \\ 
&\sim  \chi^2_{N-p-1}.
\end{align} 
It also follows immediately that $(N-p-1)\mathbb{E}\lk \hat \s^2\rk = \s^2$. Having found the distribution of $\hat \s^2$, one might consider:
\begin{align}  
\ff{\hat \b_j - \b^h_j}{\hat \s \sqrt{\left(\inv{\bX^T \bX}\right)_{ii}}} = \ff{\ff{\hat \b_j - \b^h_j}{\s \sqrt{\left(\inv{\bX^T \bX}\right)_{ii}}}}{\ff{\hat \s}{\s}}
\end{align} 
One sees that numerator of the RHS has the standard normal distribution while denominator of the RHS has the distribution $\sqrt{\frac{\chi^2_{N-p-1}}{N-p-1}}$. Furthermore, the numerator and the denominator are independent. This can be seen as follows. The projection of $\bm \e$ to the part spanned by $\ti \e_{p+2},\cds,\ti \e_N$ is given by 
$\bm{O}^T \left( \bm 1 - \bm{\ti H}\right) \bm{\ti \e} = \left( \bm 1- \bm H \right) \bm \e$. As $\hat \b_j - \b^h_j = \lk \inv{\left( \bX^T \bX \right)}\bX^T \bm \e\rk_{i}$, the contribution of $\ti \e_{p+2},\cds,\ti \e_N$ to $\hat \b_j - \b^h_j$ is given by, $\left[ \inv{\left( \bX^T \bX \right)}\bX^T  \left( \bm 1 - \bH \right)\e\right]_{i}$. However, $\bX^T \left( \bm 1 - \bm H \right) = \bR^T \bQ^T \left( \bm 1 - \bH \right) = 0$, therefore, $\hat \b_i - \b^h_i$ is a linear combination of $\ti \e_1,\cds,\ti \e_{p+1}$. On the other hand $\hat \s^2$ only involves $\ti\e_{p+2},\cds,\ti\e_N$, therefore the numerator and denominator are independent in the above equation. Hence, it follows that:
\begin{align} 
\ff{\hat \b_j - \b^h_j}{\hat \s \sqrt{\left(\inv{\bX^T \bX}\right)_{ii}}} \sim t_{N-p-1},
\end{align} 
where $t_{N-p-1}$ is the $t$ distribution with $N-p-1$ degrees of freedom. 

Another question is to ask whether one some of the features are extra. Consider the hypothesis that $\b_i = 0$ for $i = p-k+1,\cds,p$ where $1 \leq k \leq p$. Under this assumption, that is by fitting only using $\b_i$ for $k < i \leq p$. While the RSS is given by $RSS = \sum_{a=p+2}^N \left( \ti \e_a\right)^2$ in the normal setting, under the hypothesis the RSS will be given by $RSS' = \sum_{a=p-k+2}^N \left( \ti \e_a\right)^2$ (for an appropriate choice of coordinates). Therefore, one may consider
\begin{align}  
\ff{\ff{RSS'-RSS}{k}}{\ff{RSS}{N-p-1}} &= \ff{\ff{1}{k}\sum_{a=p-k+2}^{p+1}\left( \ti \e_a\right)^2}{\ff{1}{N-p-1}\sum_{a=p+2}^N \left( \ti \e_a\right)^2}.
\end{align} 
Furthermore, it is clear that numerator and denominator are independent, therefore the distribution is $F$ distribution with $k$ and $N-p-1$ degrees of freedom:
\begin{align}   
\ff{\ff{RSS'-RSS}{k}}{\ff{RSS}{N-p-1}} &\sim \ff{\ff{\chi^2_k}{k}}{\ff{\chi^2_{N-p-1}}{N-p-1}} \\ &\sim F_{k,N-p-1}. 
\end{align} 

One may also test hypothetical values of $\b$ simultaneously. That is, the hypothesis is $\bm \b = \bm \b^h$. Under this hypothesis 
\begin{align} 
\left( \bm{\hat \b} - \bm \b^h \right)^T \bX^T \bX \left( \bm{\hat \b} - \bm \b^h \right)  &= \bm \e^T \bH \bm \e \\
&= \sum_{a=1}^{p+1} \left( \ti \e_a \right)^2.
\end{align}
Therefore it follows that
\begin{align} 
\ff{N-p-1}{p+1}\ff{\left( \bm{\hat \b} - \bm \b^h \right)^T \bX^T \bX \left( \bm{\hat \b} - \bm \b^h \right)}{RSS}  &= \ff{p+1}{N-p-1}\ff{\sum_{a=1}^{p+1} \left( \ti \e_a \right)^2}{\sum_{a=p+2}^{N} \left( \ti \e_a \right)^2} \\
&\sim F_{p+1,N-p-1}.
\end{align} 

Now consider simple linear regression. That is, the model is given by $Y = \a + \b X + \e$. In this case, we denote the data by a single column vector $\bm x$. $\hat \beta$ is given by:
\begin{align} 
\hat \a &= \ff{1}{\ev{\bx,\bx}}\left( \ev{\bx,\bx}\bar y - \ev{\bx,\by} \bar x \right)
\\
\hat \b &= \frac{\ev{\bx,\by} }{\ev{\bx,\bx}}.
\end{align}
It is appropriate to introduce $R^2$ at this point, the coefficient of determination. Coefficient of determination is a measure of how much the model explains the variance in the output. $R^2 \leq 1$ and larger $R^2$ values are better than lower ones (normally $R^2 \geq 0$ as one assumes that the fit is doing better than the just using the average of the output). $R^2$ is defined as:
\begin{align} 
R^2 = 1 - \frac{\sum_{i=1}^N \left( \hat y_i - y_i \right)^2}{\sum_{i=1}^N \left( \bar y - y_i \right)^2}
\end{align} 
Let us calculate $R^2$ for simple linear regression. It is easy to see from above equations that 
\begin{align} 
\d \hat y_i &\equiv \hat y_i - y_i \\ 
&= -\d y_i + \ff{\ev{\bx,\by} }{\ev{\by,\by}} \d x_i,
\end{align} 
where $\d y_i \equiv y_i - \bar y$ and $\d x_i \equiv x_i - \bar x$. It then follows immediately that
\begin{align} 
R^2 = \frac{\ev{\bx,\by}^2}{\ev{\bx,\bx} \ev{\by,\by} }.
\end{align} 
Above, we had found that $\hat \beta_{Y \sim X} = \frac{\ev{\bx,\by} }{\ev{\bx,\bx}}$, where $Y \sim X$ means regressing $Y$ onto $X$. One may then, regress $X$ onto $Y$ and one observes that:
\begin{align} 
R^2 = \hat \b_{Y \sim X} \hat \b_{X \sim Y}.
\end{align} 
We note that while the above relation is correct for simple linear regression, it is not correct in general, i.e. for $p>1$.

Now, we consider the case of multiple linear regression. First we note that,  a simplification occurs when features are orthogonal in the sample, that is when $\bm x_i \cdot \bm x_j = 0$ for $i \neq j$. In this situation the formula for $\hat{\bm{\b}}$ is simply given by:
\begin{align} 
\hat \beta_i = \frac{\ev{\bm x_i,\by}}{\ev{\bm x_i,\bm x_i}}.
\end{align}
The idea is then to build an orthogonal set of vectors and while doing this is to calculate $\hat{\bm \b}$ "on the fly". The algorithm is given as follows. We start with $\bm z_0 = \bm x_0$ and $\hat \b^z_0 = \frac{\ev{\bm z_0,\by}}{\ev{\bm z_0, \bm z_0}}$. Them for $j= 1,\cds,p$; we define:
\begin{align} 
\bm z_j &= \bx_j - \sum_{r=0}^{j-1} \frac{\ev{\bm z_r,\bm x_j}}{\ev{\bm z_r,\bm z_r}}\bm z_r \\ 
\hat \b^z_j &= \frac{\ev{\bm z_j,\by}}{\ev{\bm z_j, \bm z_j}}.
\end{align} 
One immediately notes that, in the prediction vector $\sum_{j=0}^{N}\hat{\b}^z_j \bm z_j $, the term proportional to $\bx_p$ is comes from the term $\hat{\b}^z_p \bm z_p$ where $\bm z_p = \bm x_p + \cds$. Therefore, one finds:
\begin{align} 
\hat \b_p = \hat \b^z_p 
\end{align} 
This result can be interpreted as follows, $\hat \b_j$ is the contribution of the part of the $\bx_j$ that is orthogonal to the rest of $\bx_{i \neq j}$ to $\bm y$. Notice that one also has
\begin{align} 
Var(\hat \b_j) = \frac{\s^2}{\ev{\bm z_j,\bm z_j}}.
\end{align} 
This means that, when residual vector $\bm z_j$ is small, estimates $\hat \b_j$ will have large uncertainty. Since, small $\bm z_j$ means $\bm x_j$ is highly correlated with rest of the features, this conclusion is correct for all features that is correlated with $\bm x_j$. It follows that, the $Z$ scores for all highly correlated features will be small, while it might turn out that the $F$ score corresponding do dropping all features is large (this depends on $\bm y$, while $Z$ scores don't).

We now show how to algorithmically calculate $\hat \beta_j$. First of all we note that, $\bm X \hat{\bm \beta} = \bm Z \hat{\bm \b}^z$. In the algorithm defined above where $\bm z_j$ are calculated, let us define $\bm M_{rj} = \frac{\ev{\bm z_r,\bm x_j}}{\ev{\bm z_r,\bm z_r}}$, where $\bm M_{rj} \neq 0$ only for (we define $\bm M_{jj} =0$ for $0 \leq j \leq p$) $0 \leq r < j$ and therefore is nilpotent. It then follows that, $\bm X = \bm Z \left( \bm I +  \bm M\right) $. Therefore, we find:
\begin{align} 
\hat{\bm \b} = \left( \bm I + \bm M \right)^{-1} \hat{\bm \b}^z .
\end{align} 
Since $\bm M$ is nilpotent, it follows that
\begin{align} 
\left( \bm I + \bm M \right)^{-1} = \bm I - \bm M + {\bm M}^2 - {\bm M}^3 + \cds + \left( -1 \right)^{p} \bm M^{p}.
\end{align} 
Therefore:
\begin{align} 
\hat{\bm \b} = \left( \sum_{r=0}^{p} \left( -1\right)^{r} {\bm M}^r \right) \hat{\bm \b}^z.
\end{align}

One of the main issues with the linear regression is that when some inputs are highly correlated $\hat \b$s may have very large variance, if $\bm x_1$ is very highly correlated with $\bm x_2$ one may have $\abs{\hat {\bm \beta}_{1,2}} \gg 1$ and $\hat {\bm \b}_1 + \hat {\bm \b}_2 \approx 0$. Another issue is the fact that, when there are too many features it is difficult to interpret the results of the regression. One way to deal with these issues is called "Best Subset Selection". In best subset selection, given a number of features, among all subsets of given size the one that minimizes training error is chosen.

```python
class BestSubset():
    def __init__(self,k):
        self.k = k
    def constant_mat(self,X):
        return np.array([1.0 for _ in X.values[:,0]]).reshape(-1,1)
    def fit(self,X,y):
        RSS_arr = []
        fit_min = float('Inf')
        self.features = X.columns
        min_subset = None
        best_model = None
        for subset in itertools.combinations(self.features,self.k):
            subset = list(subset)
            if subset:
                model = LinearRegression().fit(X[subset],y)
                ssr = np.sum((model.predict(X[subset])-y)**2)
            else:
                model = LinearRegression().fit(self.constant_mat(X),y)
                ssr = np.sum((model.predict(self.constant_mat(X))-y)**2)
            if ssr < fit_min:
                fit_min = ssr
                min_subset = subset 
                best_model = model
            RSS_arr.append(ssr)
        self.subset = min_subset 
        self.model = best_model
        self.RSSs = np.array(RSS_arr)
        return self
    def predict(self,X):
        if self.subset:
            return self.model.predict(X[self.subset])
        else:
            return self.model.predict(self.constant_mat(X))
```

Below we solve the result of the best subset selection below where red points correspond to the best subset found for the given subset size. 

```python
min_arr = []
for i in range(len(features)+1):
    fit = BestSubset(i).fit(train_X,train_y)
    RSSs = fit.RSSs
    plt.scatter([i]*len(RSSs),RSSs,color='gray')
    min_arr.append(min(RSSs))
plt.plot(range(len(features)+1),min_arr,'.-.',color='red',zorder=2,ms=12)
plt.ylim(0.0,100.0);
plt.title('Best Subset')
plt.ylabel('Residual sum of squares');
plt.xlabel('# of features');
```

Notice that with the best subset selection, the larger subset size will necessarily do better. A way around this is to perform k-fold cross-validation for all subset sizes. One then finds the subset size with the minimum average RSS and then applies the chooses the subset size for which the average RSS is within one standard deviation of the minimum average RSS. Then, after finding this optimum subset size, one runs best subset algorithm for this subset size. Below we implement this process.

```python
class KFoldScan():
    def __init__(self,k=10,shuffle = True, random_state = 5, regressor = None):
        self.params = []
        self.k = k
        self.shuffle = shuffle 
        self.random_state = random_state
        self.regressor = regressor
    def add_params(self,params):
        if isinstance(params,list):
            self.params.extend(params)
        else:
            self.params.append(params)
        return self
    def scan_param(self,param,X,y):
        arr = []
        for train_idx, test_idx in self.folder():
            train_vals = X.iloc[train_idx]
            test_vals = X.iloc[test_idx]
            train_y_vals = y.iloc[train_idx]
            test_y_vals = y.iloc[test_idx]
            model = self.regressor(param).fit(train_vals,train_y_vals)
            ypred = model.predict(test_vals)
            arr.append(np.mean((ypred-test_y_vals)**2))
        self.param_dict[param] = {'RSS':np.array(arr),'mean':np.mean(arr),'std':np.std(arr)/np.sqrt(len(arr))}
    def run(self,X,y):
        self.param_dict = {}
        self.folder = lambda: KFold(n_splits = self.k, shuffle = self.shuffle,random_state=self.random_state).split(X)
        for param in self.params:
            self.scan_param(param,X,y)
        return self
    def get_RSSs(self):
        return [self.param_dict[param]['RSS'] for param in self.params] 
    def get_means(self):
        return [self.param_dict[param]['mean'] for param in self.params]
    def get_stds(self):
        return [self.param_dict[param]['std'] for param in self.params]
    def plot(self,title = 'Regression',xlabel = 'x',x_func = lambda x:x):
        x = np.vectorize(x_func)(self.params)
        means = self.get_means()
        stds = self.get_stds()
        argmin = np.argmin(means)
        ind = 0
        while means[ind]>means[argmin]+stds[argmin]:
            ind += 1
        plt.title(title)
        plt.errorbar(x,means,fmt='.',ms=8,linestyle='-.',color='red',ecolor='blueviolet',yerr = stds,capsize=5)
        plt.axvline(x=x[ind],linestyle=(0,(7,7)),color='green',linewidth=1.25)
        plt.axhline(y=means[argmin]+stds[argmin],linestyle=(0,(7,7)),color='green',linewidth=1.25)
        plt.xlabel(xlabel)
        plt.ylabel('CV Error');
```

```python
BS_fold = KFoldScan(regressor=BestSubset,k=10).add_params([i for i in range(9)]).run(train_X,train_y)
```

Another improvement on the OLS is to add a regulator term to the minimization problem. A common regulator is the Ridge regulator, where the minimization problem is given by
\begin{align} 
\hat {\bm \b}^{ridge} = \argmin_{\bm \b} \lk \ff 1 2  \left( \by - \bm X \bm \b \right)^T \left( \by - \bX \bm \b\right) + \l \sum_{i=1}^p { \b}_i^2 \rk.
\end{align} 

There is an equivalent formulation of this problem: 
\begin{align} 
\hat {\bm \b}^{ridge} &= \argmin_{\bm \b} \lk \left( \by - \bm X \bm \b \right)^T \left( \by - \bX \bm \b\right)\rk \\ 
\sum_{i=1}^p \b_i^2 &\leq t,
\end{align}
for some $t$ for each given $\l$. To see this, assume that we solved the problem with the first formulation and obtained $\hat {\bm \b}^{ridge}(\l)$ and $t(\l) \equiv \sum_{i=1}^p \hat \beta_{i}^2(\l)$. We first show that $t(\l) \leq t(0)$. Define $\mathcal{L}(\hat \b)$ to be Lagrangian (negative of the log-likelihood) of the unregularized problem. Then we know that
\begin{align} 
\mathcal{L}\left( \hat \b(0)\right) &\leq \mathcal{L} \left( \hat \b(\l) \right) \\
\mathcal{L}\left( \hat \b(\l)\right) + t(\l) &\leq \mathcal{L}\left( \hat \b(0) \right) + t(0).
\end{align} 
From these two equations it follows that $t(\l) \leq t(0)$. Therefore, we proved that the minimum of the Lagrangian does not belong to the region defined by $\sum_{i=1}^p \b_i^2 < t(\l)$. Therefore, minimizing the Lagrangian under the condition $\sum_{i=1}^p \b_i^2 \leq t(\l)$ is equivalent to the minimizing the Lagrangian under the condition $\sum_{i=1}^p \b_i^2 = t(\l)$. Now the $\l$ term doesn't vary along the surface defined by $\sum_{i=1}^p \b_i^2 = t(\l)$ and therefore minimization of the two Lagrangians restricted on this surface becomes equivalent. For the other direction, we note that for $t > t(0)$ the constraint doesn't change anything and the problem is equal to $\l=0$. For $t < t(0)$, $\l(t)$ can be found by inverting $\l \to t(\l)$.

To obtain the solution to ridge problem we first note that $\l$ doesn't multiply $\b_0^2$. It turns out that it is useful to centralize each input: 
\begin{align} 
\hat {\bm \b}^{ridge} =  \argmin_{\bm \b} \sum_{i=1}^N \left( y_i - \b_0 - \sum_{j=1}^p \bar x_j \b_j - \sum_{j} (x_{ij}-\bar x_j)\b_j \right)^2 + \l \sum_{i=1}^p \b_j^2,
\end{align}
where $\bar x_j = \ff{1}{N}\sum_{i}x_{ij}$. It follows immediately that
\begin{align} 
\hat{\beta}^{ridge}_0 + \sum_{j=1}^p \bar x_j\hat{\beta}^{ridge}_j = \bar y. 
\end{align} 
Now from $\bm X$ we remove the first column and centralize the remaining columns, that is $\bm X$ is now a $N \times p$ matrix with sum of components of each column is $0$. Similarly, we remove the first element from $\bm \b$ and also centralize $\bm y$. With these conventions, the ridge solution is given by
\begin{align} 
\hat {\bm \b}^{ridge} = \left( \bm X^T \bm X + \l \bm I \right)^{-1} \bm X^T \bm y
\end{align} 

To get more insight into what ridge regression does, let us consider the SVD decomposition of $\bm X$. Now, consider $\bm X^T \bX$, which is symmetric and positive semi-definite. We note $\bX {\bm v}_i$ is an eigenvector of $\bX^T \bX$ with eigenvalue of $\rho_i^2$. Now consider the eigenspace $E_{\bX^T \bX,\rho_i^2}$ of $\bm X^T \bm X$ with the eigenvalue $\rho_i^2$ and similarly $E_{\bX \bX^T,\rho_i^2}$. We have:
\begin{align} 
\bX &:E_{\bX^T \bX,\rho_i^2} \to  E_{\bX \bX^T,\rho_i^2} \\ 
\bX^T &:E_{\bX \bX^T,\rho_i^2} \to E_{\bX^T \bX,\rho_i^2}.
\end{align} 
These maps are isomorphisms for $\rho_i^2 \neq 0$. This follows from the fact that $\bX^T \bX = \rho_i^2 \neq 0$ on $E_{\bX^T \bX,\rho_i^2}$ and $\bX \bX^T = \rho_i^2 \neq 0$ on $E_{\bX \bX^T,\rho_i^2}$ as can be checked easily. Therefore, there is a one-to-one mapping between the eigenvalues of $\bX^T \bX$ and $\bX \bX^T$ with non-zero eigenvalues. Consider the diagonalization of $\bm X^T \bX$
\begin{align} 
\bm X^T \bX = \sum_{i=1}^p \r^2_i {\bm v}_i {\bm v}_i^T,
\end{align} 
where we assume $\rho_i \geq 0$ and ${\bm v}_i^T {\bm v}_j = \d_{ij}$. It follows that, defining $\bm u_i = \ff{1}{\rho_i}{\bX \bm v_i}$ for $\rho_i \neq 0$ and choosing orthonormal vectors for $\rho_i = 0$, one obtains:
\begin{align} 
\bX = \sum_{i=1}^p \rho_i \bm u_i \bm v_i^T.
\end{align} 
This follows from writing $\bX = \sum_{i=1}^p \bX {\bm v}_i {\bm v}_i^T$. Note that $\rho_i \bm u_i = \bX v_i$ holds for all $\rho_i$ as $\bX {\bm v}_i =0$ when $\rho_i =0$. Furthermore one may write $\bX$ in the form:
\begin{align} 
\bX = \bm U \bm D {\bm V}^T,
\end{align} 
where $\bm U$ is a $N \times p$ matrix whose columns are $\bm u_i$, $\bm D$ is a $p \times p$ diagonal matrix whose entries are $\rho_i$ and $\bm V$ is a matrix whose columns are $\bm v_i$. Next it immediately follows that from what have down above that:
\begin{align} 
\bm X \hat {\bm \b} &= \sum_{i=1}^p {\bm u_i}{\bm u_i^T}\bm y \\ 
\bm X \hat {\bm \b}^{ridge}&= \sum_{i=1}^p {\bm u_i}\frac{\rho_i^2}{\rho_i^2+\l}{\bm u_i}^T \bm y.
\end{align} 
The first equation just says that the OLS estimator projects $\bm y$ to span of $\bX$. The second equation says that with ridge estimator this components of this projection is reduced by the factor $\frac{\rho_i^2}{\rho_i^2+\l}$. When columns of $\bX$ are not linearly independent some of $\rho_i$ will vanish. Therefore, we see that when $\bX$ is close to become singular, ridge estimator reduces the projection to the singular component. 

More generally, let us order the eigenvalues and eigenvectors such that $\rho_1 \geq \rho_2 \geq \cds \rho_p \geq 0 $. Then, $\bX \bm v_1$ is the first principal component of $\bX$. Note that $\bX \bm v_1 = \rho_1 \bm u_1$ and therefore $\bm u_1$ is called the normalized first principal component of $\bm X$. Note that $\bX$ was centered, therefore $\ff{1}{N} \bX^T \bX$ is the sample covariance matrix. Therefore, $\bX \bm v_1$ is projection of the data to the direction in the feature space that has the highest variance as $Var\left( \bX \bm v_i \right) = \ff{\rho_i^2}{N}$. It follows that ridge estimator reduces the components along the directions in the feature space with small variances more than directions with large variance. This tells ridge regularization can be thought of a smooth version of PCA regression. Note that, $OLS$ projects to span of $\bX$, however small the projection of the data in a direction. A reasonable assumption is that, if a feature (principal component) is important, the variance of this feature won't be small and therefore it makes sense reduce effect of features with small variances.

Effective degrees of freedom $df(\l)$ is defined as the trace of the operator $\by \to \bX \hat{\bm \b}^{ridge}$:
\begin{align} 
df(\l) &= \Tr \left( \sum_{i=1}^p \bm u_i \frac{\rho_i^2}{\rho_i^2+\l}{\bm u}_i^T \right) \\ 
&= \sum_{i=1}^p \frac{\rho_i^2}{\rho_i^2+\l }
\end{align}

Below we find optimum value for $\l$ (or equivalently degree of the freedom) through 10-fold cross-validation and one standard deviation rule.

```python
def binary_search(xmin,xmax,func,threshold=10**(-5.0)):
    """ Finds the root of a function func in [xmin,xmax] assuming it is monotically increasing in [xmin,xmax]. """
    while xmax-xmin > threshold:
        xmed = (xmax+xmin)/2.0 
        if func(xmed) < 0:
            xmin = xmed 
        else:
            xmax = xmed 
    return (xmax+xmin)/2.0
```

```python
def ridge_fit(X,y,l,intercept=True):
    return sk.linear_model.Ridge(alpha=l,fit_intercept=intercept).fit(X,y)
```

```python
def dof(X,l):
    return sum([x**2/(x**2+l) for x in PCA().fit(X).singular_values_])
```

```python
def find_ridge(X,d,threshold=10.0**(-5)):
    l_min = 0.0 
    l_max = sum(PCA().fit(X).singular_values_**2)/d
    return binary_search(l_min,l_max,lambda l:d-dof(X,l),threshold=threshold)
```

Another regularization scheme is the Lasso regression. With Lasso regression, the minimization problem is given as:
\begin{align} 
\hat {\bm \b}^{lasso} = \argmin_{\bm \b} \lk \ff 1 2 \left( \by - \bm X \bm \b \right)^T \left( \by - \bX \bm \b\right) + \l \sum_{i=1}^p \abs{\b_i} \rk.
\end{align} 
Note that regulator term is non-analytic at $\beta_i = 0$. This has the important consequence that for large enough $\l$ some of the coefficients will be exactly zero (all coeffients will be zero for very large $\l$). This can be seen as follows. The derivative of the lagrangian with respect to $\b_i$ at $\bm{\b}$ is given by:
\begin{align} 
\frac{\partial \mathcal{L}}{\partial \beta_i} =  \lk \by^T \bX\rk_i + \l \mathop{\mathrm{sign}} \lp \b_i \rp
\end{align}
We see that, for $\l > \abs{\lk \by^T \bX \rk_i}$ as $\beta_i \to 0^+$ the derivative is positive and as $\beta_i \to 0^-$ the derivative is negative. Threfore, $\beta = 0$ is a local minimum if $\l > \abs{\lk \by^T \bX \rk_i}$ for all $i$ . Since the Lagrangian is convex, this minimum is the global minimum. Therefore, as $\l$ is increased, the coefficients will be set to $0$ one by one.

Below we will perform the Lasso fit for the prostate cancer data. A useful parameter for the Lasso regression is given by
\begin{align}
s(\l) = \frac{ \sum_{i=1}^p \abs{ \hat{\b}^{lasso}_i} }{ \sum_{i=1}^p \abs{\hat{\b}^{OLS}_i}},
\end{align}
and is called the shrinkage factor.

```python
def lasso_fit(X,y,l,intercept=True):
    if l == 0.0:
        return LinearRegression(fit_intercept=intercept).fit(X,y)
    else:
        return Lasso(alpha=l,fit_intercept=intercept).fit(X,y)
```

```python
def t(X,y,l):
    t = np.sum(np.abs(lasso_fit(X,y,l).coef_))
    return t
```

```python
def s(X,y,l):
    return t(X,y,l)/t(X,y,0.0)
```

```python
def find_lasso(X,y,s,threshold=10.0**(-5)):
    l_min = 0.0
    l_max = np.max(np.transpose(X) @ y)
    t0 = t(X,y,0.0)
    return binary_search(l_min,l_max,lambda l:s - t(X,y,l)/t0,threshold=threshold)
```

```python
ss = [k/10.0 for k in range(11)] 
lasso_params = [find_lasso(train_X,train_y,_) for _ in ss];
```

```python
L_fold = KFoldScan(regressor=Lasso,k=10).add_params(lasso_params).run(train_X,train_y)
```

```python
L_fold.plot(title='Lasso Regression',xlabel='Shrinkage Factor',x_func = lambda l: s(train_X,train_y,l))
```
